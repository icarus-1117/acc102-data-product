```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Using 'Agg' backend to prevent errors in terminal-only environments
import matplotlib
matplotlib.use('Agg')

print("=" * 60)
print("Stock Risk and Return Analysis - Server Version")
print("=" * 60)

# --- 1. Simulated Data Generation ---
print("Generating simulated stock data...")
stocks = ['AAPL', 'MSFT', 'TSLA']
dates = pd.date_range(start='2020-01-01', end='2025-01-01', freq='B')
np.random.seed(42)

initial_prices = {'AAPL': 75, 'MSFT': 160, 'TSLA': 28}
data = pd.DataFrame(index=dates)

for stock in stocks:
    params = {'AAPL': (0.0008, 0.018), 'MSFT': (0.0006, 0.015), 'TSLA': (0.0012, 0.035)}
    mu, sigma = params[stock]
    log_returns = np.random.normal(mu, sigma, len(dates))
    data[stock] = initial_prices[stock] * np.exp(np.cumsum(log_returns))

# --- 2. Metric Calculations ---
print("Calculating financial metrics...")
returns = data.pct_change().dropna()
ann_return = returns.mean() * 252
ann_risk = returns.std() * np.sqrt(252)
summary = pd.DataFrame({
    'Return': ann_return, 
    'Risk': ann_risk, 
    'Sharpe': ann_return / ann_risk
})

# --- 3. Plotting Price Trends ---
print("Plotting Price Trends...")
plt.style.use('ggplot')
plt.figure(figsize=(12, 6))

for stock in stocks:
    plt.plot(data.index, data[stock], label=f"{stock} (Final: ${data[stock].iloc[-1]:.2f})")

plt.title('Simulated Stock Price Trends (2020-2025)', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.tight_layout()

path1 = "stock_trends.png"
plt.savefig(path1, dpi=120)
print(f"File Saved Successfully: {path1}")
plt.close()

# --- 4. Plotting Risk-Return Analysis ---
print("Plotting Risk-Return Analysis...")
plt.figure(figsize=(10, 7))
scatter = plt.scatter(summary['Risk'], summary['Return'], c=summary['Sharpe'], cmap='RdYlGn', s=200, edgecolors='black')

for i in range(len(summary)):
    plt.annotate(summary.index[i], (summary['Risk'].iloc[i], summary['Return'].iloc[i]), 
                 xytext=(10, 10), textcoords='offset points', fontweight='bold')

plt.title('Risk-Return Tradeoff Analysis', fontsize=14, fontweight='bold')
plt.xlabel('Annualized Volatility (Risk)')
plt.ylabel('Annualized Return')
plt.colorbar(scatter, label='Sharpe Ratio')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

path2 = "risk_return_analysis.png"
plt.savefig(path2, dpi=120)
print(f"File Saved Successfully: {path2}")
plt.close()

# --- 5. Final Output ---
print("\nCorrelation Matrix of Returns:")
print(returns.corr().round(3))
print("\n" + "=" * 60)
print("Process Complete. Please download the PNG files to view them.")
print("=" * 60)
```
